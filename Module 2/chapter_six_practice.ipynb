```python
secret = 2
guess = 2

if guess < secret:
    print("Too low.")
elif guess > secret:
    print("Too high.")
else:
    print("Just right!")
```

```python
small = True
green = True

if not small and not green:
    print("It's a pumpkin.")
elif not small and green:
    print("It's a watermelon.")
elif small and not green:
    print("It's a cherry.")
elif small and green:
    print("It's a pea.")
```
